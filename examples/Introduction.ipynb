```python
dfs = cp.DataFramesHolder(path='../Data')
```

```python
dfs['ETC']
```

# Selection with condition

```python
dfs.drop_data_frames(key=lambda x: len(x) < 100)
```

# Preprocess Coins

```python
# Func. takes two input open and close and takes the mean of them and outputs price
dfs.preprocess({'func': 'mean', 'input': ['open', 'close'], 'output': 'price'})
# Select only price column
dfs.select_col(['price'])
```

```python
print(dfs)
```

# Merge Coins

```python
dfs.merge_frames(['ETH', 'BTC', 'ADA', 'ETH', 'XLM', 'UNI'], how='inner')
```

# Select Interval

```python
dfs.select_interval(start="2021-03-25", end="2021-04-20")
```

```python
dfs.normalize()
dfs.plot()
```
