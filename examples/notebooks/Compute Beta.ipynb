```python
import coinpy as cp
```

```python
# Load and Preprocess
dfs = cp.DataFramesHolder(path='../Data')
dfs.drop_data_frames(key=lambda x: len(x) < 1000)
dfs.preprocess({'func': 'mean', 'input': ['open', 'close'], 'output': 'price'})
dfs.sort_frames()
dfs.select_col(['price'])
dfs.dropna()
```

# Compute BETA of XLM and ETH on defined interval

```python
# Based on hourly returns
_=dfs.beta(x='XLM', y='ETH', plot=True, interval='12T') # Hourly
```

```python
# Based on daily returns
_=dfs.beta(x='XLM', y='ETH', plot=True, interval='D') 
```

```python
# Returns based on weekly returns
_=dfs.beta(x='XLM', y='UNI', plot=True,interval='W')
```
